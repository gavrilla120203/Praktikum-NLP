#**Load Dataset Menjadi Dataframe**

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# Memuat dataset dari file CSV
df = pd.read_csv('/content/data.csv')
df
```

**Interpretasi output :**

**Dataset ini terdiri dari 1000 baris dan 4 kolom yaitu "unnamed", "userName", "content", dan "score". Kolom "unnamed" adalah nomor baris dalam DataFrame yang diberikan secara default. Kolom "userName" berisikan nama pengguna dari individu yang memberikan ulasan tentang game. Kolom "content" berisikan ulasan atau komentar yang diberikan oleh pengguna tentang pengalaman mereka dengan game. Kolom "score" skor atau peringkat yang diberikan oleh pengguna untuk game tersebut.**

#**Mengubah kolom "content" menjadi vector menggunakan TFIDFVectorizer dan menampilkan hasilnya**

```python
# Inisialisasi TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()

# Mengubah kolom "content" menjadi vektor TF-IDF
tfidf_matrix = tfidf_vectorizer.fit_transform(df['content'])
```

```python
# Menampilkan DataFrame vektor TF-IDF
print(tfidf_matrix)
```

**Interpretasi output :**

**(0, 3251)	0.1055299758616986 mengindikasikan bahwa kata atau fitur yang terindeks pada posisi 3251 dalam vektor TF-IDF memiliki bobot TF-IDF sekitar 0.1055299758616986 dalam dokumen dengan nomor indeks 0.**

**(997, 1364)	0.06318278374092187 mengindikasikan bahwa kata atau fitur yang terindeks pada posisi 1364 dalam vektor TF-IDF memiliki bobot TF-IDF sekitar 0.06318278374092187 dalam dokumen dengan nomor indeks 999. Demikian seterusnya.**

#**Menampilkan nama fiturnya mengunakan get_features_name_out**

```python
# Mendapatkan daftar nama fitur
feature_names = tfidf_vectorizer.get_feature_names_out()

# Menampilkan daftar nama fitur
print(feature_names)
```

```python
# Membuat DataFrame dari vektor TF-IDF dengan nama fitur
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=feature_names)
# Menampilkan DataFrame vektor TF-IDF
print("Matriks TFIDF:")
print(tfidf_df)
```

**Interpretasi hasil :**

**Dalam matriks ini, terdapat 1000 baris yang mewakili 1000 dokumen atau ulasan berbeda dan 4353 kolom yang mewakili 4353 kata atau token yang diidentifikasi dalam dataset.**

**Dalam matriks ini, sebagian besar nilai adalah 0.0, yang menunjukkan bahwa kata-kata tersebut tidak muncul dalam dokumen yang sesuai dengan dokumen tersebut (nilai TF-IDF = 0).**

```python
# Membuat DataFrame dari vektor TF-IDF dengan nama fitur dan indeks sesuai dengan konten
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=feature_names, index=df['content'])
# Menampilkan DataFrame vektor TF-IDF
print("Matrik TFIDF:")
print(tfidf_df)
```

I**nterpretasi hasil :**

**Dari output di atas, terlihat seluruh nilai dalam matriks TF-IDF adalah 0.0 untuk setiap kata atau token dalam setiap dokumen yang ditunjukkan. Ini menunjukkan bahwa dataset yang digunakan mungkin tidak mengandung kata-kata atau token yang ada dalam daftar nama fitur.**
