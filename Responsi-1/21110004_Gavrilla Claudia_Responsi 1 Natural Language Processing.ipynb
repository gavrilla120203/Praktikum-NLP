```python
import pandas as pd
import re
```

```python
df = pd.read_excel('/content/Makassar.xlsx')
```

```python
# Fungsi untuk membersihkan teks
def bersihkan_teks(teks):
    # Hapus tautan
    teks = re.sub(r'http\S+', '', teks)

    # Hapus tanda pagar (hashtag) dan isinya
    teks = re.sub(r'#\w+', '', teks)

    # Hapus stiker dengan pola tertentu (misalnya "[stiker]")
    teks = re.sub(r'\[\w+\]', '', teks)

    # Hapus tanda baca, angka, dan karakter yang tidak penting
    teks = re.sub(r'[^\w\s]', '', teks)

    return teks
```

```python
df['full_text'] = df['full_text'].apply(bersihkan_teks)
df['tweet_url'] = ''
df['conversation_id_str'] = ''
df['user_id_str'] = ''
df['id_str'] = ''
```

```python
df.to_excel('Makassar_Bersih.xlsx', index=False)
```

#**Nomor 3 Bagian B**

```python
import pandas as pd

# Baca dataset hasil parsing
df = pd.read_excel('/content/Hasil_Parsing_Makassar_Bersih.xlsx')

# Baca file 'stopwords.csv' ke dalam DataFrame
stopwords_df = pd.read_csv('stopwords.csv')

# Ambil daftar stopwords dari DataFrame
stopwords = stopwords_df['Slangwords'].tolist()

# Buat list untuk menyimpan kata-kata slang yang ditemukan
slang_words = []

# Cari kata-kata slang
for kata in df['Kata']:
    if kata in stopwords:
        slang_words.append(kata)

# Ambil 1000 kata slang pertama (atau sesuaikan jumlahnya)
slang_words = slang_words[:1000]

# Simpan kata-kata slang dalam DataFrame
slang_df = pd.DataFrame({'Slang Words': slang_words})

# Simpan hasil dalam file Excel
slang_df.to_excel('Kata_Slang.xlsx', index=False)
```

#**Nomor 3 Bagian C**

```python
import pandas as pd

# Baca dataset hasil parsing
df = pd.read_excel('/content/Hasil_Parsing_Makassar_Bersih.xlsx')

# Baca daftar kata slangword
slang_df = pd.read_excel('/content/Kata_Slang.xlsx')
slang_words = slang_df['Slang Words'].tolist()

# Fungsi untuk tokenisasi berdasarkan kata-kata slang
def tokenisasi_dengan_slang(teks):
    tokens = teks.split()
    tokenized = []
    i = 0
    while i < len(tokens):
        token = tokens[i]
        if token in slang_words:
            # Gabungkan token dengan kata slang selanjutnya (jika ada)
            while i + 1 < len(tokens) and tokens[i + 1] in slang_words:
                token += ' ' + tokens[i + 1]
                i += 1
            tokenized.append(token)
        else:
            tokenized.append(token)
        i += 1
    return tokenized

# Terapkan fungsi tokenisasi pada seluruh kolom teks dalam dataset
df['Tokenized Text'] = df['full_text'].apply(tokenisasi_dengan_slang)

# Simpan hasil tokenisasi dalam file Excel
df.to_excel('Hasil_Tokenisasi.xlsx', index=False)
```

#**Nomor 3 Bagian D**

```python
import pandas as pd
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

# Baca dataset hasil tokenisasi
df = pd.read_excel('Hasil_Tokenisasi.xlsx')

# Fungsi untuk menghapus stopwords
def remove_stopwords(tokens):
    stop_words = set(stopwords.words('indonesian'))
    tokens = [token for token in tokens if token.lower() not in stop_words]
    return tokens

# Terapkan fungsi penghapusan stopwords pada kolom 'Tokenized Text'
df['Tokenized Text'] = df['Tokenized Text'].apply(remove_stopwords)

# Simpan hasil penghapusan stopwords dalam file Excel
df.to_excel('Hasil_Stopword_Removal.xlsx', index=False)
```

```python
import pandas as pd
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# Baca dataset hasil penghapusan stopwords
df = pd.read_excel('Hasil_Stopword_Removal.xlsx')

# Inisialisasi stemmer Sastrawi
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# Fungsi untuk melakukan stemming pada token
def stemming(tokens):
    return [stemmer.stem(token) for token in tokens]

# Terapkan fungsi stemming pada kolom 'Tokenized Text'
df['Tokenized Text'] = df['Tokenized Text'].apply(stemming)

# Tampilkan 100 stemming pertama
first_100_stems = df['Tokenized Text'].head(100).explode().unique()

# Simpan hasil stemming dalam file Excel
df.to_excel('Hasil_Stemming.xlsx', index=False)

# Tampilkan 100 stemming pertama
print(first_100_stems)
```
